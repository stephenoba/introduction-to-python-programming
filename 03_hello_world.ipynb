# Hello World

It is a tradition in any programming language to get started with writting a *Hello World* style program. In Python the simplest version of the *Hello World* program merely prints out a string with the message ***Hello World***

```python
# Run this using any of the methods of running a python program
print('Hello World')
```

The `print()` function is an example of an inbuilt function and comes by default with Python. It can be used to print things out. The output is actually printed to what is known as the output stream. This handles a stream (sequence) of data such as letters and numbers. This output stream of data can be sent to an output window such as the terminal on
a Mac or Command Window on a Windows PC. In this case we are printing the string 'Hello World'.

The `print()` function actually tries to print whatever you give it. If it is given a string like `"Hello World"` it will print out a string, if it is given an integer like `42` it will print out that integer, similarly, if it is given a floating point number such as `3.14` it will print out that same floating point number.

We can have our *Hello WOrld* program written in a file and ran on the command line (on windows) or the Teerminal (on Mac and Linux).
1. Create a file named `hello.py` and open it with your IDE of choice (like pycharm)
2. Type the line `print("Hello World")` in the file you created, then save.
3. Open your Command Line or Terminal and make sure you are in the same directory as your `hello.py` file.
4. Type `python hello.py` into the Command line or terminal.

Let's stretch our *Hello World* program a little bit by making it interactive. We would get our program to ask us our name and favourite color and give us a personalized message. Update your `hello.py` file to look like this.

```python
print('Hello World')
user_name = input("Enter your name: ")
color = input("Enter your favourite color: ")
print(f"Hello {user_name}, {color} is a beautiful color!")
```

In our *Hello World* program, we added three new statements. The last statement we have seen before. but the second and third statements are new and similar. Let's look at this statement.

The `user_name = input("Enter your name: ")` first executes another inbuilt function `input()`. This function is passed a string which is known as an argument. The argument `Enter your name: ` is used when the `input()` function promts the user for an input. 

Whatever the user types after being prompted for an input is then stored in a variable called `user_name`.

A variable is a named area of the computersâ€™ memory that can be used to hold data such as strings, numbers, boolean such as True/False etc

This simplified picture illustrates how a variable, such as user_name, can reference an area of memory containing actual data. In this diagram the memory is shown as a two dimensional grid of memory locations. Each location has an address associated with it. This address is unique within the memory and can be used to return to the data held at that location. This address is often referred to as the memory address of the data. It is this memory address that is actually held in the variable user_name; this is why the user_name variable is shown as pointing
to the area in memory containing the string 'John'. Thus the variable user_name allows us to access this area of memory easily and conveniently. For example, if we want to get hold of the name entered by the user in another statement, we can do so merely by referencing the variable user_name.
