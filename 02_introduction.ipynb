# Quick Introduction to Python

### What is Python?

Simply put by wikipedia, Python is an interpreted high-level general-purpose programming language.

Guido van Rossum started working on it in the 1980s and was first released in 1991. The language is named after one of Guido's favorite TV programs “Monty Pythons Flying Circus”. You will even find various references to this show in the official Python documentation

Python has become popular for the following factors;

1. Its flexibility and simplicity which makes it easy to learn.
2. Its use by the Data Science community where it provides a more standard
programming language than some rivals such as R.
3. Its suitability as a scripting language for those working in the DevOps field
where it provides a higher level of abstraction than alternative languages tra-
ditionally used.
4. Its ability to run on (almost) any operating system, but particularly the big three
operating systems Windows, MacOS and Linux.
5. The availability of a wide range of libraries (modules) that can be used to extend
the basic features of the language.
6. It is free!

### Python Programming

There are different programming paradigms that a programming languages may allow developers to code in like, Procedural programming, declarative programming, functional programming, and object oriented programming.

Python is considered to be a hybrid language; that is it allows developers to utilise a combination of different approaches within the same program. It allows you to write very procedural code, to use objects in an object oriented manner and to write functional programs

### Python Libraries

libraries extend the functionality of a language and make it much easier to develop applications and programs. Python has lot of these libraries availabe. Some of which are;

1. Frameworks like Django and Flask for the web
2. smtplib (a SMTP email client) and imaplib (an IMAP4email client) which are email clients

4. Python Excel Libraary for generating Microsoft excel files

### Python Execution Model

As we have already defined earlier, python is not a precompiled language in the way that some other languages you
may have come across are such as C++. Instead it is an interpreted language. An interpreted language is one that does not require a separate compilation phase to convert the human readable format into something that can be executed by a computer. Instead the plain text version is fed into another program (generally referred to as the
interpreter) which then executes the program for you.

### Running Python Programs

There are different ways to run a python program. It could be run by;

* **Interactively using the python interpreter** - This is possible with the use of the Python REPL (named after *R*ead *E*valuate *Print* *L*oop style of operation). With the REPL statements can me typed directly and immediately executed. To run the REPL python must first be installed. The REPL can be initiated from a Command Prompt on Windows and from the terminal on Mac and Linux distributions.

* **Through the Python IDLE** - The Python IDLE (Integrated Development and Learning Environment) is an Integrated development environment (IDE) for Python. It comes with the Windows installer by default. On linux, it needs to be installed. This can be done by running the command `sudo apt-get installl idle`

* **Runing a python file on the comand line or terminal using the `python` comand** - We can store Python commands into a file. This creates a program file that can then be run as an argument to the `python` command. The file must be named with the `.py` extension.

* **Running as a script file** - We can transform a file containing a python program into a script. A script is a stand alone file that can be run directly without the need to explicitly use the `python` comand. This is done by adding a special line to the start of the Python file that indicates the Python command (or interpreter) to use with the rest of the file.
